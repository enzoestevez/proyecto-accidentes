```python
# Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Leer el archivo cargado
df = pd.read_csv('accidentes_transito_argentina.csv')

# Mostrar primeras filas
df.head()

```

```python
# Info general del dataset
df.info()

# Estadísticas descriptivas
df.describe()

# Provincias con más accidentes
df['Provincia'].value_counts()
```

```python
# Ver cantidad de accidentes por provincia
df['Provincia'].value_counts().plot(kind='bar')
plt.title('Accidentes por provincia')
plt.show()
```

```python
# Gráfico de causas más frecuentes
sns.countplot(data=df, x='Causa', order=df['Causa'].value_counts().index)
plt.title('Causas de accidentes')
plt.xticks(rotation=45)
plt.show()

```

```python
# Fallecidos por tipo de accidente
df.groupby('TipoAccidente')['Fallecidos'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Fallecidos por tipo de accidente')
plt.ylabel('Cantidad de fallecidos')
plt.show()
```

Conclusión:
El análisis muestra que las provincias con más accidentes son Buenos Aires y Córdoba. Las causas más frecuentes fueron el exceso de velocidad y la distracción. Los tipos de accidentes con mayor cantidad de fallecidos fueron los choques frontales y vuelcos.
